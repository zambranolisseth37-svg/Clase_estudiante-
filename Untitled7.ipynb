```python
class Estudiante:
    def __init__(self, nombre, apellido, cedula, email, semestres, materias):
        self.__nombre = nombre
        self.__apellido = apellido
        # Usamos los setters para validar desde la creación
        self.cedula = cedula
        self.__email = email
        self.semestres = semestres
        self.__materias = materias

    # --- GETTERS ---
    @property
    def cedula(self): return self.__cedula

    @property
    def semestres(self): return self.__semestres

    # --- SETTERS CON VALIDACIÓN ---
    @cedula.setter
    def cedula(self, valor):
        # Convertimos a string para verificar longitud y tipo
        valor_str = str(valor)
        if valor_str.isdigit() and len(valor_str) == 10:
            self.__cedula = valor_str
        else:
            print("Error: son 10 números")
            self.__cedula = None # Opcional: manejar el estado de error

    @semestres.setter
    def semestres(self, valor):
        if isinstance(valor, int) and 1 <= valor <= 9:
            self.__semestres = valor
        else:
            print("Error: solo número del 1 al 9")
            self.__semestres = None

    # --- MÉTODO __STR__ DINÁMICO ---
    def __str__(self):
        # self.__dict__ captura todos los atributos del objeto automáticamente
        return str(self.__dict__)

# --- Ejemplo de uso ---

# Caso correcto
estudiante1 = Estudiante("Juan", "Pérez", "1234567890", "juan@mail.com", 5, ["Matemáticas", "Física"])
print(estudiante1)

# Caso con error en cédula (ponemos 5 números en lugar de 10)
estudiante2 = Estudiante("Ana", "López", "123", "ana@mail.com", 2, ["Historia"])

# Caso con error en semestre (ponemos 12 en lugar de 1 al 9)
estudiante3 = Estudiante("Luis", "Gomez", "0987654321", "luis@mail.com", 12, ["Química"])

# Demostración de adición dinámica de atributos
estudiante1.promedio = 9.5
print(f"Estudiante 1 con nuevo atributo: {estudiante1}")
```
